```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Rm = 1
RM = 2
x = []
y = []
for _ in range(1000):
    r = np.sqrt(np.random.rand()*(RM*RM-Rm*Rm) + Rm*Rm)
    theta = 2*np.pi*np.random.rand()
    x.append(r*np.cos(theta))
    y.append(r*np.sin(theta))
    
plt.scatter(x,y)
```
